## Data Cleaning, Data Wrangling

### 1.Identify and handle missing values

- A missing value could be represented as ?, N/A or just a blank cell

```python
# To replace missing value indicating objects with NaN, use the following command
df1=df.replace('?' ,np.NaN)
df1.isnull().sum()
```

```python
df1["b"].value_counts(dropna= False ) # counts the number of each type of values in column "b"
```

```python
#df.replace({'old':"new" }, inplace= True) #can be used to replace object with any other value in the dataset (*not missing values)
```

```python
#df.dropna(subset=["feature"],axis=0, inplace= True) #to drop missing columns
                                                    #axis=0, drops entire row
                                                    #axis=1, drops entire column
```

```python
df1.dropna(subset=["z","b","s","t","f"], axis=0, inplace= True)   #Removed the rows with missing values in column "z"
df1.isnull().sum()
```

```python
df1.info
```

### 2.Data Formatting

#### Datatype Conversions

```python
df1.dtypes
```

```python
df1["b"]
```

```python
df1["b"]=df1["b"].astype("int")
```

#### Formatting Columns

```python
df1["b"]=10/df1["b"]
print(df1.head())
```

#### Rename Columns

```python
# Renaming columns
df1.rename(columns={"a":"aa"}, inplace= True)
df1.info()
```

### 3.Data Normalization(centering/ scaling)

- simple feature scaling

```python
df1["x"]= df1["x"]/df1["x"].max()
df1["x"]
```

- Min- Max

```python
df1["y"]= (df1["y"]-df1["y"].min())/(df1["y"].max()-df1["y"].min())
df1["y"]
```

- Z- Score

```python
df1["z"]= df1["z"].astype(int)  #convert to data type int before using normalization
df1["z"]= (df1["z"]-df1["z"].mean())/df1["z"].std()
df1["z"].std()
```

### 4.Data Binning

```python
binwidth= int((max(df1["n"])-min(df1["n"]))/4)
bins= range(min(df1["n"]), max(df1["n"]), binwidth)
group_names= ['low', 'medium', 'high']
df1["n-binned"]= pd.cut(df1["n"], bins, labels= group_names)
#error
```

```python
bins=[1488, 2000, 3000, 4066]
group_names= ["low","good","great"]
df1["n-binned"]= pd.cut(df1["n"], bins, labels= group_names)
df1["n-binned"]
```
